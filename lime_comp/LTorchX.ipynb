```python
X, y  = sklearn.datasets.make_moons(n_samples=10000, noise= 0.2)
```

```python
## Train a back-box classifier on the synthetic dataset
import sklearn.model_selection, sklearn.ensemble

train, test, labels_train, labels_test = sklearn.model_selection.train_test_split(X, y, test_size=0.20)

clf = sklearn.ensemble.RandomForestClassifier(n_estimators=200)
clf.fit(train, labels_train)
print(sklearn.metrics.accuracy_score(labels_test, clf.predict(test)))
print(sklearn.metrics.classification_report(labels_test, clf.predict(test)))
```

```python
explainer = lime.lime_tabular.LimeTabularExplainer(X, discretize_continuous = False)
```

```python
def get_LIME_coeffs(points, explainer):
    output = []
    for i, point in enumerate(points):
        exp = explainer.explain_instance(point, clf.predict_proba, num_features=2, top_labels=1)
        class_ = clf.predict([point])
        intercept = list(exp.intercept.values())[0] #There's an irritating edge case where class_ =/= the class in the exp.
        coefs = [x[1] for x in sorted(exp.as_list(list(exp.intercept.keys())[0]), key=lambda x:x[0])]
        
        coefs.append(intercept)
        output.append(coefs)
        
    return output
```

```python
C = get_LIME_coeffs(points,explainer)
```

```python
def plot_explanations(coefficients, points):
    palette = sns.color_palette("bright", n_colors=len(coefficients))
    
    mask = y[:1000] == 0

    plt.scatter(X[:1000,0][np.nonzero(y[:1000])],X[:1000,1][np.nonzero(y[:1000])], c='lightgray', marker='x', label='Class 0')
    plt.scatter(X[:1000,0][mask],X[:1000,1][mask], c='gray', marker='x', label='Class 1')

    for i, coeff_ in enumerate(coefficients):
        intercept = coeff_[2]

        x1s = [np.min(X), np.max(X)]
        y1s = [((0.5 - intercept - x1*coeff_[0])/coeff_[1]) for x1 in x1s]

        plt.plot(x1s, y1s, color=palette[i], linestyle=':', linewidth=4)
        plt.scatter(points[i][0],points[i][1], color=palette[i],marker='8', linewidth=4)
   
    axes = plt.gca()

    axes.set_xlim([np.min(X, axis=0)[0], np.max(X, axis=0)[0]])
    axes.set_ylim([np.min(X, axis=0)[1], np.max(X, axis=0)[1]])
    plt.show()
```

```python
def forward(points, pred_plus, variat):
    
    sc = Sample_Concrete(2,2,batch_size = points.shape[0])
    
    logits = pred_plus(points)
    samples = sc(logits)    
    new_ps = samples * points
    preds = variat(new_ps)
    
    return preds
```

```python
def get_L2X_coeffs(points, clf, pred_plus, variat): #For the 0.5 boundary point (note that this is often too noisy to be useful - I've added the clf boundary as a simpler check)
    coeffs = []
    points = torch.tensor(points)
    
    weights = pred_plus(points.float())
    
    for i in range(weights.shape[0]):
        
        point = points[i,:]
        weight = weights[i,:].double()
        
        top = point - 2. * weight #Only go in direction of (neg) weight
        #bottom = point - weight

        a = torch.linspace(top[0].data[0],point[0].data[0], steps=1000)
        b = torch.linspace(top[1].data[0],point[1].data[0], steps=1000)

        pointlist = torch.zeros((1000,2))
        pointlist[:,0] = a
        pointlist[:,1] = b
        
        #plt.scatter(a, b, linewidth=0.2)
        #plt.scatter(point[0].data[0],point[1].data[0], linewidth=4)
        
        problist1 = forward(pointlist, pred_plus, variat).detach().numpy()
        problist2 = clf.predict_proba(pointlist)
        
        #x = range(len(problist1))
        #plt.plot(x, problist1.detach().numpy()[:,0])
        #plt.plot(x, problist2[:,0])

        neglist1 = np.abs(problist1[:,0]-problist1[:,1])
        neglist2 = np.abs(problist2[:,0]-problist2[:,1])

        #Bias according to NN
        idx1 = np.argmin(neglist1)
        cross_point1 = pointlist[idx1,:]
        Y_val1 = problist1[idx1,0]
        
        weight=weight.detach().numpy() #For float-double issue reasons
        
        bias1 = Y_val1 - weight[0]*cross_point1[0] - weight[1]*cross_point1[1]
        bias1 = bias1.detach().numpy()
        
        #Bias according to clf
        
        idx2 = np.argmin(neglist2)
        cross_point2 = pointlist[idx2,:]
        Y_val2 = problist2[idx2,0]

        bias2 = Y_val2 - weight[0]*cross_point2[0] - weight[1]*cross_point2[1]
        bias2 = bias2.detach().numpy()
        
        coeffs.append([[weight[0], weight[1], bias1.item(0)], [weight[0], weight[1], bias2.item(0)]])
    
    return coeffs
```

```python
def get_crap_L2X_coeffs(points, clf, pred_plus, variat): #For the point itself
    coeffs = []
    points = torch.tensor(points)
    
    weights = pred_plus(points.float())
    
    for i in range(weights.shape[0]):
        
        point = points[i,:]
        weight = weights[i,:].double()
        cross_point1 = point.detach().numpy()
        
        weight=weight.detach().numpy() #For float-double issue reasons
        
        bias1 = 0.5 - weight[0]*cross_point1[0] - weight[1]*cross_point1[1]
        
        coeffs.append([weight[0], weight[1], bias1.item(0)])
    
    return coeffs
```

```python
from sklearn.metrics import roc_curve, auc
from scipy.special import expit
```

```python
def lin_pred(point, coefs):
    pred = expit(coefs[2] + np.dot(point, coefs[:2]))
    return pred
```
